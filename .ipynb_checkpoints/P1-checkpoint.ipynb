Below are some helper functions to help get you started. They should look familiar from the lesson!

````python
import math
import numpy as np

buf_mPos = np.array([0.,0.,0.,0.]) ; buf_xcPos = np.array([0.,0.,0.,0.]) ; buf_ycPos = np.array([0.,0.,0.,0.])
buf_mNeg = np.array([0.,0.,0.,0.]) ; buf_xcNeg = np.array([0.,0.,0.,0.]) ; buf_ycNeg = np.array([0.,0.,0.,0.])
wma_weights = np.array([0.4,0.3,0.2,0.1]) # Weights for 4 sample weighted moving average
smooth_lines = True

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image


def draw_lines(img, lines, color=[255, 0, 0], thickness=10):
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
    global buf_mPos, buf_xcPos, buf_ycPos, buf_mNeg, buf_xcNeg, buf_ycNeg, wma_weights, smooth_lines
    shp = img.shape # get shape to determine the bounds for extrapolating line
    # Define slopes (pos & neg) and points (center, top & bottom)
    mPos = 0 ; mNeg = 0 ; mPosCount = 0; mNegCount =0 ;
    mPosPoints = 0 ; mNegPoints = 0
    xcPos = 0 ; ycPos = 0 ; xbPos = 0 ; ybPos = shp[0] ; xtPos = 0 ; ytPos = 0.6*shp[0]
    xcNeg = 0 ; ycNeg = 0 ; xbNeg = 0 ; ybNeg = shp[0] ; xtNeg = 0 ; ytNeg = 0.6*shp[0]
    for line in lines:
        for x1,y1,x2,y2 in line:
            m = (y2-y1)/(x2-x1)
            # Process only if line is between 10-80 degrees, skip otherwise
            # From camera, perspective lines for lanes are unlikely to be out of this range
            # The average centroid of all lines and slope is stored for extrapolation later
            if abs(m) < 0.5 or abs(m) > 5.0:
                continue
            if m > 0:
                mPos += m ; mPosCount += 1
                xcPos += (x1+x2)/2 ; ycPos += (y1+y2)/2
            if m < 0:
                mNeg += m ; mNegCount += 1
                xcNeg += (x1+x2)/2 ; ycNeg += (y1+y2)/2
    if mPosCount==0 or mNegCount==0:
        return
    mPos /= mPosCount
    xcPos /= mPosCount
    ycPos /= mPosCount
    mNeg /= mNegCount
    xcNeg /= mNegCount
    ycNeg /= mNegCount
    
    if smooth_lines:
        buf_mPos = np.roll(buf_mPos,1) ; buf_mPos[0] = mPos ; mPos = np.dot(buf_mPos,wma_weights)/np.sum(wma_weights[buf_mPos!=0])
        buf_xcPos = np.roll(buf_xcPos,1) ; buf_xcPos[0] = xcPos ; xcPos = np.dot(buf_xcPos,wma_weights)/np.sum(wma_weights[buf_xcPos!=0])
        buf_ycPos = np.roll(buf_ycPos,1) ; buf_ycPos[0] = ycPos ; ycPos = np.dot(buf_ycPos,wma_weights)/np.sum(wma_weights[buf_ycPos!=0])
        buf_mNeg = np.roll(buf_mNeg,1) ; buf_mNeg[0] = mNeg ; mNeg = np.dot(buf_mNeg,wma_weights)/np.sum(wma_weights[buf_mNeg!=0])
        buf_xcNeg = np.roll(buf_xcNeg,1) ; buf_xcNeg[0] = xcNeg ; xcNeg = np.dot(buf_xcNeg,wma_weights)/np.sum(wma_weights[buf_xcNeg!=0])
        buf_ycNeg = np.roll(buf_ycNeg,1) ; buf_ycNeg[0] = ycNeg ; ycNeg = np.dot(buf_ycNeg,wma_weights)/np.sum(wma_weights[buf_ycNeg!=0])
    
    # Extend line to ends of the ROI as defined
    xbPos = xcPos + (1/mPos)*(ybPos-ycPos)
    xtPos = xcPos + (1/mPos)*(ytPos-ycPos)
    xbNeg = xcNeg + (1/mNeg)*(ybNeg-ycNeg)
    xtNeg = xcNeg + (1/mNeg)*(ytNeg-ycNeg)
    #print('xcPos=',xcPos,'ycPos=',ycPos,'xcNeg=',xcNeg,'ycNeg=',ycNeg,'mPos=',mPos,'mNeg=',mNeg)
    #print('xbPos=',xbPos,'ybPos=',ybPos,'xtPos=',xtPos,'ytPos=',ytPos)
    #print('xbNeg=',xbNeg,'ybNeg=',ybNeg,'xtNeg=',xtNeg,'ytNeg=',ytNeg)
    cv2.line(img, (int(xbPos), int(ybPos)), (int(xtPos), int(ytPos)), color, thickness)
    cv2.line(img, (int(xbNeg), int(ybNeg)), (int(xtNeg), int(ytNeg)), color, thickness)

def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    draw_lines(line_img, lines)
    return line_img

def draw_curves(img, lines, color=[255, 0, 0], thickness=10):
    from numpy import polyfit
    from numpy import poly1d
    shp = img.shape # get shape to determine the bounds for extrapolating line
    ptxL = [] ; ptyL = [] ; ptxR = [] ; ptyR = []
    for line in lines:
        for x1,y1,x2,y2 in line:
            m = (y2-y1)/(x2-x1)
            if abs(m) < 0.5 or abs(m) > 5.0:
                continue
            if m > 0:
                ptxL.append(x1) ; ptxL.append(x2)
                ptyL.append(y1) ; ptyL.append(y2)
            if m < 0:
                ptxR.append(x1) ; ptxR.append(x2)
                ptyR.append(y1) ; ptyR.append(y2)
    # Fit x as a function of y !!!
    if len(ptxL)==0 or len(ptxR)==0:
        return
    coeffL = polyfit(ptyL, ptxL, 2)
    polyL = poly1d(coeffL)
    coeffR = polyfit(ptyR, ptxR, 2)
    polyR = poly1d(coeffR)
    # Generate points across ROI extent
    ys = np.linspace(0.6*shp[0], 1.0*shp[0], 100)
    xsL = polyL(ys)
    xsR = polyR(ys)
    
    # Draw polygonal curves
    crvL = np.array([])
    crvR = np.array([])
    for i in range(len(ys)):
        crvL = np.append(crvL,[xsL[i],ys[i]])
        crvR = np.append(crvR,[xsR[i],ys[i]])
    crvL = np.reshape(crvL,(-1,2),order='C')
    crvR = np.reshape(crvR,(-1,2),order='C')
    
    #img_test = np.zeros((100,100,3), np.uint8)
    #pts = np.array([[10,5],[20,30],[70,20],[50,10]], np.int32)
    #img_test = cv2.polylines(img_test,[pts],True,(0,255,255))
    #cv2.imshow('test',img_test)
    #cv2.waitKey(0)
    #cv2.destroyAllWindows()
    
    #img_test = np.zeros((shp[0],shp[1],3), np.uint8)
    # See http://stackoverflow.com/questions/17241830/opencv-polylines-function-in-python-throws-exception
    #cv2.polylines(img_test,np.int32([crvL, crvR]),False,(0,0,255))
    #cv2.imshow('test',img_test)
    #cv2.waitKey(0)
    #cv2.destroyAllWindows()
    
    cv2.polylines(img,np.int32([crvL, crvR]),False,color=(0,0,255),thickness=10)

def hough_curves(img, rho, theta, threshold, min_line_len, max_line_gap):
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    curve_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    draw_curves(curve_img, lines)
    return curve_img

# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, =0.8, =1., =0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img *  + img *  + 
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, , img, , )
````

### Implementation Overview

The project was completed by implementing the image processing pipeline that utilized the starter functions. The pipeline performs the following steps

1. Load frame from video and converts to grayscale
2. Apply gaussian blur and identify edges using Canny routine 
3. Creates a region of interest to __*mask non-relevant*__ edges (trees, other cars, etc).
4. Uses masked edges to determine the set of Hough lines
5. Hough lines are then drawn on the image and frame is output to video

The key implemention is in function __draw_lines__. The function segregates the Hough lines in the parameter space and allocates them to either the left or the right lanes depending on the apparent slope. The implementation provides a flag __smooth_lines__, which performs *Simple Weighted Moving Average (SWMA)* smoothing of the line in parameter space. The output video without and with SWMA smoothing is presented in two separate videos. The pipeline was also successfully applied to the last video with yellow left lane.

### Key implementation

The challenge problem was also attempted. The video output using the previous implementation of __draw_lines__ resulted in poor tracking of curved lanes. Two alternate functions __draw_curves__ and __hough_curves__ were implemented to employ quadratic fit approach. However, due to poor sampling from dashed center-lanes, the quality of tracking is poor. The tracking of lanes in frames for the concrete section of the road is especially difficult. Some ideas for further improvement are as follows (these are being explored and will be implemented in due course of time).

* Improve lane detection in concrete section
  - Incresae image contrast before Canny edge detection
  - Change Gaussian blur parameter
  - Change the minVal and maxVal for Canny function
* Cubic or spline based fitting of points
* Further improvement to masked regions (two polygonal regions, one for each lane)
* Keep a history of points from the last three frames while fitting. This will improve and smoothen the fit for dashed lanes.
